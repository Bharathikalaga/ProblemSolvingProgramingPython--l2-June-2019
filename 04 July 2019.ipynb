```python
# graph to plot distance and time
distance = list(range(2,11,2))
time = list(range(1,6))
# Distance  = 2* time
plt.title("distance vs Time")
plt.xlabel("time")
plt.ylabel("distance")
plt.plot(time,distance,"bo")
plt.grid()
plt.show()
```

```python
# v = at
t = list(range(1,11))
v = [2 * i for i in range(1,11)]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
# v = at
t = list(range(1,11))
a = list(range(1,len(t)+1))
v = [a[i] * i for i in range(len(t))]
plt.plot(t,v,"g")
plt.grid()
plt.show()
```

```python
n = list(range(10))
sqn = [i * 2 for i in n]
plt.plot(n,sqn)
#plt.grid()
plt.show()
        
```

```python
# Voltage - time
import numpy as np
t = [i/10 for i in range(1,21)]

v =[ 1 + np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.grid()
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.show()
```
